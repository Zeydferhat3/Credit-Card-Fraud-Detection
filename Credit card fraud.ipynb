# 1. Imports

```python
import pandas as pd
import  numpy as np
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# 2. data

```python
data=pd.read_csv("creditcard.csv")
```

# 3. Scaling

```python
X = data.drop(columns = 'Class', axis=1)
Y = data['Class']
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
X = standardized_data
```

# 4.Splitting data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# 5.Models

## 5.1 Support Vector Machine

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
# accuracy score on the training data
X_train_pred = classifier.predict(X_train)
acc = accuracy_score(X_train_pred, Y_train)
```

```python
# accuracy score on the test data
X_test_pred = classifier.predict(X_test)
acc1 = accuracy_score(X_test_pred, Y_test)
```

```python
print('trainning accuracy  : ', acc)
print('testing   accuracy :  ', acc1)
```

## 5.2 LogisticRegression()

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# accuracy score on the training data
X_train_pred = model.predict(X_train)
acc = accuracy_score(X_train_pred, Y_train)
```

```python
# accuracy score on the test data
X_test_pred = model.predict(X_test)
acc1= accuracy_score(X_test_pred, Y_test)
```

```python
print('trainning accuracy  : ', acc)
print('testing   accuracy :  ', acc1)
```

## 5.3 Decsision Tree

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=0, max_depth=2)
```

```python
decision_tree.fit(X_train, Y_train)
```

```python
# accuracy score on the training data
X_train_pred = decision_tree.predict(X_train)
acc = accuracy_score(X_train_pred, Y_train)
```

```python
# accuracy score on the test data
X_test_pred = decision_tree.predict(X_test)
acc1 = accuracy_score(X_test_pred, Y_test)
```

```python
print('trainning accuracy  : ', acc)
print('testing   accuracy :  ', acc1)
```
